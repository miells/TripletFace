##JIT Compile the model

```
from tripletface.core.model import Encoder
import torch
```

```
model = Encoder(64)
weights = torch.load("model2/model.pt")['model']
model.load_state_dict( weights )
jit_model = torch.jit.trace( model, torch.rand(8, 3, 3, 3) )
torch.jit.save( jit_model, "/content/TripletFace/model2/scriptModel.pt" )
```
